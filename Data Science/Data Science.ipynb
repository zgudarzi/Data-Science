```python
### change the names of the columns 

# Creating the DataFrame
df = pd.DataFrame(data)

# Replace spaces with underscores in column names
df.columns = df.columns.str.replace('\xa0', '__')
df.columns = df.columns.str.replace(' ', '_')

# Print the updated column names to verify
print("Updated column names:")
print(df.columns)

```

```python

for col in df.columns:
    df[col] = df[col].replace(['> 1000'], "1001.0")
df
```

```python
#for filling the missing values of " 'Quality__of_Faculty'"
df3= df.dropna()
plt.scatter(df3['World_Rank'], df3['Quality__of_Faculty'])
```

```python

# Replace '-' with NaN
df.replace('-', np.nan, inplace=True)

# Convert columns to float
df['Quality__of_Education'] = df['Quality__of_Education'].astype(float)
df['Quality__of_Faculty'] = df['Quality__of_Faculty'].astype(float)

# Fill NaN values with the maximum value plus one
df['Quality__of_Education'].fillna(df['Quality__of_Education'].max() + 1, inplace=True)
df['Quality__of_Faculty'].fillna(df['Quality__of_Faculty'].max() + 1, inplace=True)

print(df)
```

```python
# Converting to float except columns of  'Institution', 'Location'
columns=df.columns 

for i in columns:
    if (df[i].dtype!= 'float64') and  ( i != "Location") and (i != "Institution") :
        df[i] = df[i].astype ("float" )
column_types = df.dtypes
print(column_types)
```

```python

import geopandas as gpd
import matplotlib.pyplot as plt
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter



# Function to get geographical coordinates
geolocator = Nominatim(user_agent="geoapiExercises")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

def get_coordinates(row):
    try:
        location = geolocator.geocode(f"{row['Institution']}, {row['Location']}")
        return pd.Series([location.latitude, location.longitude])
    except:
        return pd.Series([None, None])

# Apply the function to get coordinates for each university
df[['Latitude', 'Longitude']] = df.apply(get_coordinates, axis=1)

# Drop rows with missing coordinates
df = df.dropna(subset=['Latitude', 'Longitude'])

# Load the world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create a GeoDataFrame for the universities
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))

# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1)
world.plot(ax=ax, color='lightgrey')

# Plot universities
gdf.plot(ax=ax, color='red', markersize=20)

# Annotate universities
#for x, y, label in zip(gdf.geometry.x, gdf.geometry.y, gdf['Institution']):
   # ax.text(x, y, label, fontsize=8, ha='right', color='blue')

# Set title and show plot
ax.set_title('Universities on World Map')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the number of institutions in each location
location_counts = df['Location'].value_counts()

# Threshold for displaying portions 
threshold = 0.05

# Filter out small portions
large_portions = location_counts[location_counts / location_counts.sum() >= threshold]
small_portions = location_counts[location_counts / location_counts.sum() < threshold]

# Group small portions into 'Other'
other_count = small_portions.sum()
large_portions['Other'] = other_count

# Create a pie plot for large portions
plt.figure(figsize=(10, 8))
plt.pie(large_portions, labels=large_portions.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 12})
plt.title('Proportion of Institutions by Location')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

```
